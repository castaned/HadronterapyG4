```python
import pandas as pd
import matplotlib.pyplot as plt
```

Leemos los archivos "Dose" y lo cargamos como datasets

```python
Dose_90 = pd.read_csv("Dose_90.out", delim_whitespace=True)
Dose_70 = pd.read_csv("Dose_70.out", delim_whitespace=True)
Dose_60 = pd.read_csv("Dose_60.out", delim_whitespace=True)
Dose_50 = pd.read_csv("Dose_50.out", delim_whitespace=True)
```

Vemos el contenido que contiene y su estadística básica (con uno es suficiente)

```python
Dose_50.head()
```

```python
Dose_50.describe()
```

Renombramos la columna "i" que es la dimensión por la que se mueven los protones por "mm" medidas de distancia 

```python
Dose_60["i"] = Dose_60["i"] * 0.1
Dose_70["i"] = Dose_70["i"] * 0.1
Dose_50["i"] = Dose_50["i"] * 0.1
Dose_90["i"] = Dose_90["i"] * 0.1
Dose_60 = Dose_60.rename(columns={"i": "mm"})
Dose_70 = Dose_70.rename(columns={"i": "mm"})
Dose_50 = Dose_50.rename(columns={"i": "mm"})
Dose_90 = Dose_90.rename(columns={"i": "mm"})
```

Graficamos las 4 diferentes dosis en mismo plot para ver como se comportan

```python
plt.figure(figsize=(10, 6))

plt.plot(Dose_90['mm'], Dose_90['Dose(Gy)'], label='Dosis con 90 MeV')
plt.plot(Dose_70['mm'], Dose_70['Dose(Gy)'], label='Dosis con 70 MeV')
plt.plot(Dose_60['mm'], Dose_60['Dose(Gy)'], label='Dosis con 60 MeV')
plt.plot(Dose_50['mm'], Dose_50['Dose(Gy)'], label='Dosis con 50 MeV')

plt.xlabel('mm')
plt.ylabel('Dosis (Gy)')
plt.title('Dosis depositada en el voxel')
plt.legend()
plt.grid(True)

plt.show()
```

Vemos el pico máximo y su valor correspondiente en su longitud para cada Dosis.

```python
max = Dose_90['Dose(Gy)'].max()
value_max = Dose_90.loc[Dose_90['Dose(Gy)'].idxmax(), 'mm']
print(f'El valor máximo de Dose(Gy) es: {max}')
print(f'El valor de i correspondiente al valor máximo de Dose(Gy) es: {value_max}')
```
